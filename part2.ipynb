```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import pandas as pd
```

```python
# read training datas from seperate files and merging them
measurements_train = pd.read_csv("train\\measurements.csv")
metadata_train = pd.read_csv("train\\hwg_metadata.csv")
df_train = pd.merge(measurements_train,metadata_train, on= 'subject_id' )
df_train
```

```python
#dropping unecessary columns
df_train.drop(["height_cm","subject_id"],axis=1,inplace=True)
df_train
```

```python
#reading and merging the test data
measurements_test = pd.read_csv("test\\measurements.csv")
metadata_test = pd.read_csv("test\\hwg_metadata.csv")
df_test = pd.merge(measurements_test,metadata_test, on= 'subject_id' )
df_test
```

```python
# again drop unnecessary columns
df_test.drop(["height_cm","subject_id"],axis=1,inplace=True)
df_test
```

```python
# x,y test ,train data splitting 
y_train = df_train["weight_kg"].values
x_train = df_train.drop("weight_kg",axis=1)
x_train
```

```python
y_train
```

```python
y_test= df_test["weight_kg"].values
x_test = df_test.drop("weight_kg",axis=1)
x_test
```

```python
y_test
```

```python
#Preprocess data encoding gender column
from sklearn.preprocessing import LabelEncoder
label_encoder =LabelEncoder()
x_train["gender"] = label_encoder.fit_transform(x_train["gender"] )
x_train
```

```python
# encoding gender column for test data
x_test["gender"] = label_encoder.fit_transform(x_test["gender"] )
x_test
```

```python
#scalling train and test x values w,th standar scaler
from sklearn.preprocessing import StandardScaler
myscaler = StandardScaler()
x_train_scaled = myscaler.fit_transform(x_train)
x_test_scaled = myscaler.transform(x_test)
x_train_scaled
```

```python
x_test_scaled
```

```python
#Fit a Model without PCA (no reduction)
from sklearn.linear_model import LinearRegression

model_no_reduction = LinearRegression()
model_no_reduction.fit(x_test_scaled,y_test)
```

```python
#calculate 3 metrices of linear model without pca
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
preds_no_reduction = model_no_reduction.predict(x_test_scaled)

print("MSE (without reduction): ", mean_squared_error(y_test,preds_no_reduction))
print("MAE (without reduction): ", mean_absolute_error(y_test,preds_no_reduction))
print("R2 (without reduction): ", r2_score(y_test,preds_no_reduction))
```

```python
# implement pca with different k values, fitting a model and calculating metrices of this model iteratively
from sklearn.decomposition import PCA
mse_values = []
mae_values = []
r2_scores = []
ks =[2,4,6,8,10]

for k in ks:
    pca = PCA(n_components=k)
    x_train_reduced = pca.fit_transform(x_train_scaled)
    x_test_reduced = pca.transform(x_test_scaled)

    model_reduction = LinearRegression()
    model_reduction.fit(x_train_reduced,y_train)

    preds_reduction = model_reduction.predict(x_test_reduced)
    mse_values.append(mean_squared_error(y_test,preds_reduction))
    mae_values.append(mean_absolute_error(y_test,preds_reduction))
    r2_scores.append(r2_score(y_test,preds_reduction))
```

```python
#plotting mse scores vs each corresponding k values graph
plt.plot(ks, mse_values, marker="o")
plt.title("MSE vs K")
plt.xlabel("K values")
plt.ylabel("MSE")
plt.show()
```
