```python
#work on only the training data.
#we will be focusing on only the independent variables. Your target is the weight.
#So, remove the target variable from the data. 

wieghts = df_train["weight_kg"]
indep_vars = df_train.drop("weight_kg",axis=1)
indep_vars
```

```python
#â€¢ Do not forget to encode the gender variable. 
from sklearn.preprocessing import LabelEncoder
label_encoder =LabelEncoder()
indep_vars["gender"] = label_encoder.fit_transform(indep_vars["gender"] )
indep_vars #1 male 0 female
```

```python
#StandScaler to a common procedure is to preprocess the data so that each feature has 0 mean and unit variance
from sklearn.preprocessing import StandardScaler
myscaler = StandardScaler()
scaled_indep_vars = myscaler.fit_transform(indep_vars)
scaled_indep_vars
```

```python
scaled_indep_vars.shape
```

```python
#covariance matrix calculation
myCov = np.cov(scaled_indep_vars.T)
```

```python
myCov.shape
```

```python
#eigenvectors and eigenvalues computation
eValues, eVectors = np.linalg.eig(myCov)
eValues
```

```python
eVectors
```

```python
#printing each eigen value and corresponding eigen vectors
for i in range(len(eValues)):
    print(f"Eigenvalue {i+1}:  {eValues[i]}")
    print(f"Corresponding Eigenvector: {eVectors[:, i]}\n")
```

```python
#creating a dataframe for each eigen value and corresponding eigen vectors for the report (format it to 2 decimals after dot for better visualization)
eigenvalues_formatted = [f'{e:.2f}' for e in eValues]
eigenvectors_formatted = [[f'{v:.2f}' for v in vec] for vec in eVectors.T]
eigen_df = pd.DataFrame({'Eigenvalues': eigenvalues_formatted, 'Eigenvectors': eigenvectors_formatted})

pd.set_option('display.max_colwidth', 1000)
eigen_df
```

```python
#eValues already sorted from highest to lowest 
#Eigenvalue 1:  9.464991000462094
#Eigenvalue 2:  2.9224609520895224
#Eigenvalue 3:  0.8486937044260575
eVectors_highest3 = eVectors[:,:3]

coef_table = pd.DataFrame(eVectors_highest3, columns = ["Component 1","Component 2","Component 3"], index=indep_vars.columns)
```

```python
# Draw the scree graph (eigenvalues vs eigenvectors)
plt.ylabel("Eigenvalues")
plt.xlabel("Component Number")
plt.title('Scree Plot')
plt.ylim()
plt.style.context('seaborn-whitegrid')
plt.axhline(y=1, color='r',linestyle='--')
plt.plot(range(1,len(eValues)+1), eValues,marker="o", linestyle='-', c='b')
```

```python
#explained variance calculation
varianceProp=eValues/eValues.sum()
varianceProp
```

```python
#cumulative axplained variance calculation
cum_var_exp=np.cumsum(varianceProp)
cum_var_exp
```

```python
#explained variance ratio vs pricipal components grapgh plotting with cumulatice explained variance line
plt.bar(range(1,len(varianceProp)+1),varianceProp, align='center',label='individual explained variance')
plt.step(range(1,len(cum_var_exp)+1),cum_var_exp, where='mid', label='cumulative explained variance', c='red')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.show()
```

```python
#For each K (number of selected components) value from 1 to M (total number of components)
#Reconstruct your data back. 
#Calculate and record the reconstruction error. 
print(indep_vars.shape)
errors = []

for k in range(1,len(eValues)+1):
    comps = eVectors[: ,:k]
    indep_vars_reduced = scaled_indep_vars.dot(comps)
    reduced_data = indep_vars_reduced.dot(comps.T)
    reconstruciton_error = np.mean(np.sum((scaled_indep_vars- reduced_data)**2,axis=1))
    errors.append(reconstruciton_error)


```

```python
#plotting reconstruction error vs k graph
plt.figure(figsize=(10,6))
plt.plot(range(1,len(eValues)+1), errors, marker ="o", color='red')
plt.title("Reconstruction Error vs K")
plt.xlabel("K values")
plt.ylabel("Reconstruction error")
plt.show()
```

```python
#Select the first 2 components (with highest eigenvalues), project the data on to the plane
x_reduced= scaled_indep_vars.dot(eVectors[:,0:2])
plt.figure(figsize=(10,6))
plt.scatter(x_reduced[:,0], x_reduced[:,1], s=20, alpha= 0.5,c="green")
plt.title("Projected Data Visualization")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```
